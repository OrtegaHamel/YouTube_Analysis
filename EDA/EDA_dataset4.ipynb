### Análisis distribución de variables numéricas

```python
# Generamos estadísticas descriptivas del Data Frame
print("=" * 50)
print("Estadísticas descriptivas del Data Frame: ")
print("Columnas numéricas:")
display(df.describe().T)
```

El resumen estadístico revela una presencia crítica de valores nulos enmascarados como valores negativos (-1 y -2) en múltiples variables de interacción y ratios (como likes/views, videoViewCount y likes/dislikes), lo que podría alterar cualquier métrica si no se realiza un filtrado o imputación previa. Por su parte, las distribuciones demuestran una asimetría positiva extrema provocada por outliers masivos típicos de la economía del entrenemiento. Al contrastar las medianas (percentil 50%) contra los valores máximos en métricas como subscriberCount (408 vs. 25.2 millones) y videoViewCount (4,807 vs. 500.8 millones), podemos afirmar que una fracción mínima de canales y videos hipervirales domina el ecosistema, distorsionando las medias aritméticas y obligando a utilizar enfoques de análisis robustos basados en rangos intercuartílicos durante la preparación de los datos.

```python
# Bloque 1: Volúmenes Absolutos
columnas_volumen = [
    'videoViewCount', 'videoLikeCount', 'videoDislikeCount', 'VideoCommentCount',
    'channelViewCount', 'channelCommentCount', 'videoCount', 'subscriberCount'
]

print("=== Asimetría: Métricas de Volumen ===")
for col in columnas_volumen:
    print(f"{col:<20} : {df[df[col] >= 0][col].skew():.4f}")

fig, axes = plt.subplots(nrows=8, ncols=2, figsize=(14, 45))
fig.suptitle('Volúmenes Absolutos de Videos y Canales (Datos >= 0)', fontsize=16, fontweight='bold', y=0.99)
sns.set_theme(style="whitegrid")

for i, col in enumerate(columnas_volumen):
    df_temporal = df[df[col] >= 0]
    
    # Histograma Logarítmico
    sns.histplot(data=df_temporal, x=df_temporal[col] + 1, kde=True, ax=axes[i, 0], color='royalblue', bins=30, log_scale=True)
    axes[i, 0].set_title(f'Distribución de {col}', fontsize=11, fontweight='semibold')
    axes[i, 0].set_ylabel('Frecuencia')
    
    # Boxplot
    sns.boxplot(data=df_temporal, x=col, ax=axes[i, 1], color='lightskyblue', fliersize=3)
    axes[i, 1].set_title(f'Outliers en {col}', fontsize=11, fontweight='semibold')

plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

Los gráficos muestran distribuciones con una asimetría positiva extrema y severa, acorde a la economía de la atención en plataformas digitales. Al aplicar la escala logarítmica en los histogramas, se evidencia que variables como videoViewCount y subscriberCount tienden a un comportamiento de distribución normal, reflejando que la gran mayoría de los videos y canales acumulan métricas concentradas en las colas bajas, mientras que los boxplots exhiben una densidad alta de valores atípicos hacia los rangos superiores. Esta dispersión masiva confirma que una minoría de creadores y canales consolidados domina el volumen bruto de interacciones del ecosistema, lo que valida la necesidad obligatoria de descartar la media aritmética como métrica de tendencia central antes del modelado supervisado.

## Tratamiento de nulos en las columnas numéricas
Revisaremos la alta presencia de valores negativos (-1 y -2) en múltiples variables de interacción y ratios que podrían ser nulos enmascarados.

```python
# Contar cuántos valores negativos hay por columna
for col in df.select_dtypes(include=[np.number]).columns:
    negativos = (df[col] < 0).sum()
    if negativos > 0:
        porcentaje = (negativos / len(df)) * 100
        print(f"Columna '{col}': {negativos} valores negativos ({porcentaje:.2f}%)")


data_errores = {
    'Columna': [
        'likes/dislikes', 'comments/subscriber', 'likes/subscriber', 'dislikes/subscriber',
        'views/subscribers', 'totviews/totsubs', 'comments/views', 'VideoCommentCount',
        'dislikes/views', 'likes/views', 'videoLikeCount', 'videoDislikeCount',
        'totvideos/videocount', 'videoViewCount', 'views/elapsedtime'
    ],
    'Porcentaje_Negativos': [
        96.50, 5.06, 4.92, 4.92, 3.11, 3.10, 2.11, 2.09, 1.99, 1.99, 1.98, 1.98, 0.05, 0.02, 0.02
    ],
    'Cantidad_Absoluta': [
        555445, 29121, 28341, 28341, 17915, 17827, 12120, 12034, 11463, 11463, 11370, 11370, 268, 104, 104
    ]
}

df_errores = pd.DataFrame(data_errores)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(11, 7))

colores = ['#e63946' if x > 90 else '#457b9d' for x in df_errores['Porcentaje_Negativos']]

barplot = sns.barplot(
    data=df_errores,
    x='Porcentaje_Negativos',
    y='Columna',
    palette=colores
)

for index, row in df_errores.iterrows():
    barplot.text(
        row['Porcentaje_Negativos'] + 1,  # Posición X (un poco a la derecha de la barra)
        index,                            # Posición Y
        f"{row['Porcentaje_Negativos']:.2f}% ({row['Cantidad_Absoluta']:,})", 
        va='center', 
        fontweight='semibold', 
        fontsize=10,
        color='#1d3557'
    )

plt.title('Incidencia de Valores Negativos (Nulos Ocultos) por Columna', 
          fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Porcentaje de Registros Afectados (%)', fontsize=11)
plt.ylabel('Variables del Dataset', fontsize=11)
plt.xlim(0, 115) 

plt.tight_layout()
plt.show()
```

Se ve claramente el caso crítico de likes/dislikes con el 96.50% de negativos. Se eliminará esta columna del DataFrame (df.drop), ya que la presencia masiva del valor -2 representa un código de error de extracción derivado de indeterminaciones matemáticas (división por cero) o de las restricciones de privacidad implementadas en 2021 por YouTube en el conteo de dislikes.  

Variables estructurales como videoViewCount y views/elapsedtime presentaron registros negativos en apenas un 0.02% de la muestra. Al tratarse de fallos en las variables raíz del rendimiento, se procederá a la remoción de estas filas, que no comprometen la representatividad estadística del volumen de datos.

Para los ratios de enganche y contadores que dieron negativo,como comments/subscriber con 5.06%, likes/subscriber con 4.92%, y likes/views con 1.99%, la lógica de negocio nos dice que el video simplemente no recibió interacciones (0 likes o 0 comentarios), o el canal no tiene suscriptores públicos. Por lo tanto, se aplicará una imputación por cero absoluto (0.0).

```python
# Remover la columna 'likes/dislikes'
df_clean = df.drop(columns=['likes/dislikes']).copy()

# Filtramos datos físicamente imposibles: las métricas de visualización y tiempo no pueden ser negativas.
df_clean = df_clean[df_clean['videoViewCount'] >= 0]
df_clean = df_clean[df_clean['views/elapsedtime'] >= 0]

# Forzamos las interacciones negativas a cero
columnas_interacciones_absolutas = ['videoLikeCount', 'videoDislikeCount', 'VideoCommentCount']
for col in columnas_interacciones_absolutas:
    df_clean[col] = df_clean[col].clip(lower=0)

# Ratios negativos normalizamos a 0
columnas_ratios_negativos = [
    'likes/subscriber', 'views/subscribers', 'dislikes/views', 
    'comments/subscriber', 'likes/views', 'comments/views', 
    'dislikes/subscriber', 'totviews/totsubs', 'totvideos/videocount'
]

for col in columnas_ratios_negativos:
    # Sustituye cualquier valor menor que cero por cero absoluto
    df_clean[col] = df_clean[col].apply(lambda x: 0.0 if x < 0 else x)

print(f"Dimensiones originales: {df.shape}")
print(f"Dimensiones tras la limpieza: {df_clean.shape}")
print(f"Total de valores negativos restantes: {(df_clean.select_dtypes(include=[np.number]) < 0).sum().sum()}")
```

```python
total_original = 575610
filas_limpias = 575506
filas_eliminadas = total_original - filas_limpias  

porcentaje_conservado = (filas_limpias / total_original) * 100
porcentaje_eliminado = (filas_eliminadas / total_original) * 100

labels = ['Registros Conservados\n(Datos Válidos)', 'Registros Eliminados\n(Inconsistencias)']
sizes = [porcentaje_conservado, porcentaje_eliminado]

sns.set_theme(style="white")
fig, ax = plt.subplots(figsize=(8, 6))

explode = (0, 0.2) 
colores = ['#457b9d', '#e63946']  

wedges, texts, autotexts = ax.pie(
    sizes, 
    explode=explode, 
    labels=labels, 
    autopct='%1.3f%%', 
    shadow=False, 
    startangle=140, 
    colors=colores,
    textprops=dict(color="#1d3557", fontweight="semibold", fontsize=11)
)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)

texto_detalle = (
    f"Detalle Físico del Dataset:\n"
    f"• Total Inicial: {total_original:,} filas\n"
    f"• Conservadas: {filas_limpias:,} filas\n"
    f"• Eliminadas: {filas_eliminadas} filas"
)
ax.text(
    1.2, -0.1, texto_detalle, 
    fontsize=10, 
    bbox=dict(boxstyle="round,pad=0.5", facecolor="#f1faee", edgecolor="#a8dadc", lw=1)
)

plt.title('Impacto del Saneamiento de Datos sobre las Observaciones\n(Filtrado de Registros Corruptos)', 
          fontsize=13, fontweight='bold', pad=20, color='#1d3557')

plt.tight_layout()
plt.show()
```

El resultado final del proceso arrojó un DataFrame consolidado de 575,506 filas y 26 columnas, reduciendo exitosamente el inventario de valores negativos a cero. Este ecosistema de datos mitiga cualquier riesgo de sobreajuste o distorsión matemática, quedando óptimamente acondicionado para las etapas de entrenamiento del modelo predictivo.

## Análisis de Correlación

```python
columnas_numericas = [
    'videoViewCount', 
    'VideoCommentCount', 
    'comments/views', 
    'views/elapsedtime'
]
df_numerico = df_clean[columnas_numericas]

matriz_correlacion = df_numerico.corr(method='spearman')
print("=== Relaciones clave de correlación (Spearman) ===")
correlaciones_ordenadas = matriz_correlacion.unstack().sort_values(ascending=False)
print(correlaciones_ordenadas[correlaciones_ordenadas < 1.0].drop_duplicates().head(5))

sns.set_theme(style="white")
plt.figure(figsize=(8, 6.5))

sns.heatmap(
    matriz_correlacion,
    annot=True, 
    fmt=".3f",  
    cmap='coolwarm',     
    vmin=-1, vmax=1, 
    linewidths=1.5,      
    square=True,         
    cbar_kws={"shrink": .7, "label": r"Coeficiente de Correlación de Spearman ($\rho$)"} 
)

plt.title('Matriz de Correlación de Spearman', 
          fontsize=13, fontweight='bold', pad=20, color='#1d3557')

plt.xticks(rotation=15, ha='right', fontsize=10, fontweight='semibold', color='#1d3557')
plt.yticks(rotation=0, fontsize=10, fontweight='semibold', color='#1d3557')

plt.tight_layout()
plt.show()
```

La matriz de Spearman revela una sinergia casi lineal y dominante de 0.950 entre videoViewCount y la velocidad de maduración temporal views/elapsedtime, lo que valida matemáticamente el uso de esta tasa como eje central de la variable objetivo, al demostrar que el volumen acumulado está encadenado a la velocidad de tracción. Asimismo, el ratio relativo comments/views registra un acoplamiento robusto de 0.761 con VideoCommentCount, mientras que este último mantiene una correlación moderada y estable frente a las métricas de visualización (views/elapsedtime de 0.554 y videoViewCount de 0.550). # Diagnóstico y Selección de Problema

A partir del análisis exploratorio, se confirma que este dataset corresponde a una cápsula de tiempo de la plataforma altamente fragmentada y dispersa, caracterizada por registrar de manera orgánica la primera década completa de vida de YouTube (desde agosto de 2005 hasta octubre de 2015). Esta distribución temporal permite analizar la maduración estructural del algoritmo de recomendación original. Estructuralmente, el dataset exhibe una granularidad extrema con 449,980 canales y 555,627 videos únicos, donde el 96.40% de la muestra (535,644 registros) aparece una sola vez y solamente el 3.60% (19,983 registros) cuenta con dos apariciones. También se detectó una anomalía crítica de calidad de datos y nulos ocultos en la variable likes/dislikes, que registró una tasa de corrupción extrema del 96.50% debido a operaciones matemáticas imposibles o restricciones de privacidad de la API (valores centinela -2), obligando a su descarte definitivo. Los valores negativos restantes se redujeron a cero mediante la remoción de inconsistencias en visualizaciones (0.02% de filas con vistas negativas) y la imputación por cero absoluto (0.0) en ratios de interacción donde el negativo representaba operacionalmente la ausencia de tracción o restricciones de metadatos comunitarios.  

Las variables numéricas del dataset exhiben asimetrías severas por el fenómeno de la Cola Larga (Long Tail), pero la matriz de Spearman revela interacciones orgánicas altamente consistentes para el modelado. Destaca una sinergia casi lineal y dominante de 0.950 entre videoViewCount y la velocidad de maduración temporal views/elapsedtime, lo que valida matemáticamente el uso de esta tasa como eje central de la variable objetivo, al demostrar que el volumen acumulado está encadenado a la velocidad de tracción. Además, el ratio relativo comments/views registra un acoplamiento robusto de 0.761 con VideoCommentCount, que mantiene una correlación moderada y estable frente a las métricas de visualización (views/elapsedtime de 0.554 y videoViewCount de 0.550). Esta independencia parcial prueba que la activación comunitaria posee una dinámica de interacción propia, consolidando al vector de comentarios como un sustituto predictivo idóneo y metodológicamente defendible para mitigar la ausencia de la polaridad que aportaban los dislikes.  

Frente a estas características, se plantea abordar un problema de Regresión y Optimización mediante un Sistema Predictivo y Prescriptivo para la Optimización de Lanzamientos en YouTube. En el ecosistema saturado de la creación de contenido, el éxito de un video depende críticamente de decisiones estratégicas de publicación tomadas antes del estreno. Las agencias de marketing, marcas y productoras necesitan herramientas científicas para simular escenarios de lanzamiento. Se diseñará una solución interactiva de software con un enfoque dual alimentado por el mismo motor de Machine Learning. El primer componente, el Predictor, devolverá una estimación cuantitativa del índice de velocidad de tracción esperado, permitiendo al creador simular si su configuración base pertenece estadísticamente al segmento de alto rendimiento o al estándar. El segundo componente, el Optimizador, actuará como un módulo prescriptivo que extraerá las fronteras de decisión y reglas de los árboles entrenados para recetar de forma automatizada las condiciones ideales de lanzamiento (día, horario y duración) para maximizar la tracción en una categoría específica.  

La variable objetivo (y) será cuantitativa continua, orientada a medir la velocidad de tracción y madurez del video combinada con el tiempo transcurrido (views/elapsedtime). Las variables predictoras (X) incluirán la componente temporal descompuesta mediante el accesor .dt en variables cíclicas discretas (año, mes, día de la semana y hora de publicación), las componentes categóricas formales de temática y las métricas estructurales de los canales. Se excluirán los identificadores únicos de alta cardinalidad (videoId) por su nulo valor predictivo, así como la columna corrupta likes/dislikes.  

## Desafíos  
El desafío técnico de este dataset radica en la escalabilidad y procesamiento ante la alta cardinalidad de creadores. Intentar aplicar técnicas tradicionales como One-Hot Encoding colapsaría la memoria RAM del sistema, lo que obligará al pipeline de preprocesamiento (ColumnTransformer) a implementar estrategias avanzadas de codificación no lineal como Target Encoding supervisado o la agrupación agregada de canales por volumen de producción. Asimismo, la gran dispersión en las métricas de interacción obligará a utilizar algoritmos de ensamble basados en árboles no paramétricos (como Random Forest, XGBoost o LightGBM) debido a su capacidad nativa para capturar interacciones no lineales complejas sin sufrir por la correlación de estructuras.El desafío metodológico exigirá calibrar minuciosamente los hiperparámetros de los modelos (como max_depth y min_samples_leaf) mediante GridSearchCV con validación cruzada (cv=5) para evitar que los regresores se sesguen por el fenómeno de la "Cola Larga". Se priorizará la optimización y evaluación diagnóstica basada en métricas de regresión robustas como el Error Absoluto Medio (MAE) y el R-cuadrado (R^2), enfocadas en minimizar la desviación del índice de velocidad estimado frente a la tracción real observada. Adicionalmente, el valor del negocio se defenderá mediante el Análisis de Importancia de Características (feature_importances_) y Gráficos de Dependencia Parcial (PDP) para aislar científicamente las curvas de rendimiento y revelar los puntos ideales empíricos de lanzamiento, consolidando así una herramienta analítica altamente confiable y explicable para la toma de decisiones comerciales.
