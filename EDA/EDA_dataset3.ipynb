### Análisis distribución de variables numéricas

```python
# Generamos estadísticas descriptivas del Data Frame
print("=" * 50)
print("Estadísticas descriptivas del Data Frame: ")
print("Columnas numéricas:")
display(df.describe())
```

Se ve una dispersión extrema y una severa asimetría positiva en todas las métricas de interacción. La brecha entre las medianas y los promedios delata la presencia de outliers masivos o fenómenos de viralidad concentrada.

```python
print("=== Coeficiente de Asimetría (Skewness) ===")
print(df[['view_count', 'like_count', 'comment_count']].skew())

columnas_num = ['view_count', 'like_count', 'comment_count']

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 22))
fig.suptitle('Data 3: Análisis de Distribución de Columnas Numéricas', fontsize=16, fontweight='bold', y=0.95)

# Establecer estilo estético profesional
sns.set_theme(style="whitegrid")

# 3. Iterar sobre las columnas para dibujar Histograma + KDE y Boxplot
for i, col in enumerate(columnas_num):
    # Columna izquierda: Histograma con curva de densidad (KDE)
    sns.histplot(data=df, x=df[col] + 1, kde=True, ax=axes[i, 0], color='royalblue', bins=30, log_scale=True)
    axes[i, 0].set_title(f'Histograma de {col}', fontsize=12, fontweight='semibold')
    axes[i, 0].set_xlabel('')
    axes[i, 0].set_ylabel('Frecuencia')
    
    # Columna derecha: Gráfico de caja (Boxplot) para identificar Outliers
    sns.boxplot(data=df, x=col, ax=axes[i, 1], color='lightskyblue')
    axes[i, 1].set_title(f'Boxplot de {col}', fontsize=12, fontweight='semibold')
    axes[i, 1].set_xlabel('')

# Ajustar los espacios para que no se solapen los títulos
plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()

```

Los coeficientes confirman una asimetría positiva (derecha) severa en todas las métricas de interacción, siendo view_count la variable más extrema con un valor de 8.00. Este comportamiento estadístico confirma que la gran mayoría de los videos se concentra en rangos de rendimiento bajos o moderados, mientras que un grupo menor de videos virales actúa como outliers masivos desplazando la distribución hacia la derecha, lo que requerirá transformaciones matemáticas obligatorias para estabilizar la varianza antes de entrenar cualquier modelo.

```python
# Identificar cuantitativamente los outliers usando el criterio IQR (1.5 * IQR)
columnas_analisis = ['view_count', 'like_count', 'comment_count']

print("=== Conteo de Outliers por Variable ===")
for col in columnas_analisis:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_superior = Q3 + 1.5 * IQR
    
    outliers = df[df[col] > limite_superior]
    porcentaje = len(outliers) / len(df) * 100
    print(f"{col}: {len(outliers)} videos atípicos ({porcentaje:.2f}% del dataset) -> Umbral: > {limite_superior:,.0f}")

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(14, 18))
fig.suptitle('Análisis Avanzado de Outliers (Escala Logarítmica)', fontsize=14, fontweight='bold', y=1.02)

sns.set_theme(style="whitegrid")

for i, col in enumerate(columnas_analisis):
    # Usamos escala logarítmica en el eje X para poder "separar" los outliers y ver su distribución
    sns.boxplot(data=df, x=col, ax=axes[i], color='salmon', flierprops={"markerfacecolor":"red", "markeredgecolor":"red", "markersize": 5})
    axes[i].set_xscale('log')
    axes[i].set_title(f'Distribución de Atípicos en {col}', fontsize=12, fontweight='semibold')
    axes[i].set_xlabel('Valor (Escala Log10)')

plt.tight_layout()
plt.show()
```

El conteo cuantitativo y la inspección visual de los diagramas de caja confirman una presencia crítica de outliers que afecta fuertemente a la muestra, destacando que hasta un 14% de los videos califica como atípico en sus niveles de conversación (comment_count > 4,487). En los boxplots, la longitud de los bigotes superiores y la densa dispersión de puntos hacia el extremo derecho demuestran que el fenómeno de la viralidad no es un evento aislado en este dataset, sino un componente estructural donde un 13.33% de los registros supera holgadamente el umbral de las 14.8 millones de vistas, inflando artificialmente las medias analíticas.  

En consecuencia, los outliers no serían eliminados, ya que corresponden a la realidad del mercado audiovisual y el funcionamiento propio de la plataforma.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import Counter
import ast

# 1. Configurar el estilo base y un lienzo unificado con dimensiones óptimas
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

# Título global del lienzo unificado
fig.suptitle('Análisis Temático y de Integridad', fontsize=15, fontweight='bold', y=0.98)

# =========================================================================
# GRÁFICO 1: Top 10 Canales Más Frecuentes
# =========================================================================
top_channels = df['channel_title'].value_counts().head(10)

# Corrección crítica: Se asigna explícitamente el gráfico al primer eje ax=axes[0]
sns.barplot(x=top_channels.values, y=top_channels.index, ax=axes[0], palette="viridis")

# Configuración formal usando métodos set_ para subplots
axes[0].set_title('Top 10 Canales Más Frecuentes en el Dataset', fontsize=12, fontweight='bold', pad=10)
axes[0].set_xlabel('Cantidad de Videos', fontsize=10)
axes[0].set_ylabel('Nombre del Canal', fontsize=10)


# =========================================================================
# GRÁFICO 2: Análisis de Dispersión y Dominancia de Tags (Top 10 Más Comunes)
# =========================================================================
lista_tags_limpias = []
for celda in df['tags'].dropna():
    try:
        # Evaluar de forma segura el string como lista si viene formateado como tal
        tags_evaluadas = ast.literal_eval(celda) if isinstance(celda, str) and celda.startswith('[') else str(celda).split(',')
        for tag in tags_evaluadas:
            tag_limpia = tag.strip().replace("'", "").replace('"', "").lower()
            if tag_limpia: # Evitar vacíos
                lista_tags_limpias.append(tag_limpia)
    except Exception:
        pass

conteo_tags = Counter(lista_tags_limpias)
df_top_tags = pd.DataFrame(conteo_tags.most_common(10), columns=['Tag', 'Frecuencia'])

# Corrección: Este ya apuntaba bien a ax=axes[1]
sns.barplot(
    data=df_top_tags, 
    x='Frecuencia', 
    y='Tag', 
    ax=axes[1], 
    palette='flare'
)

# Configuración formal usando métodos set_ para el segundo subplot
axes[1].set_title('Top 10 Frecuencias de tags', 
                  fontsize=12, fontweight='bold', pad=10)
axes[1].set_xlabel('Frecuencia de Aparición', fontsize=10)
axes[1].set_ylabel('Etiqueta (Tag)', fontsize=10)


# 2. Ajustar los espacios automáticamente para evitar solapamientos de texto
plt.tight_layout()
plt.show()
```

Se observa una concentración moderada liderada por el canal de tecnología "Mrwhosetheboss", quien posee la mayor frecuencia de participación con 27 videos asignados (4.5% del dataset). Este volumen de publicaciones por emisor sugiere la viabilidad de agrupar a los creadores de menor peso bajo una etiqueta común de "Otros" si se desea utilizar el canal como una variable predictiva categórica sin inflar la dimensionalidad del modelo. ### Análisis de Correlación

```python
columnas_numericas = ['view_count', 'like_count', 'comment_count']
df_numerico = df[columnas_numericas]

matriz_correlacion = df_numerico.corr(method='spearman')

print("=== Relaciones clave de correlación (Spearman) ===")
print(matriz_correlacion.unstack().sort_values(ascending=False).drop_duplicates().head(10))

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

sns.heatmap(
    matriz_correlacion, 
    annot=True,          
    fmt=".3f",           
    cmap='coolwarm',     
    vmin=0, vmax=1,      
    linewidths=1.5,      
    square=True,         
    cbar_kws={"shrink": .8, "label": r"Coeficiente de Correlación ($\rho$)"} 
)

plt.title('Matriz de Correlación Interacciones de Video', fontsize=13, fontweight='bold', pad=15)
plt.xticks(rotation=0, fontsize=10, fontweight='semibold')
plt.yticks(rotation=0, fontsize=10, fontweight='semibold')

plt.tight_layout()
plt.show()
```

La relación más estrecha se evidencia entre las reproducciones y la aprobación pasiva, donde view_count y like_count registran un coeficiente de Spearman de alto impacto de 0.854, demostrando que el volumen de visualizaciones actúa como el dinamizador directo de la aceptación del contenido. La conversión hacia la sección de comentarios exhibe una sutil holgura que corrobora que la interacción activa requiere un nivel de involucramiento mayor por parte del usuario, esto se refleja en coeficientes robustos pero ligeramente menores cuando comment_count se cruza con las vistas (0.743) y con los likes (0.731). En términos algorítmicos, esta estructura lineal sólida valida que el rendimiento del contenido en el dataset se rige bajo un ecosistema de tracción proporcional unificado, donde el alcance expansivo arrastra de forma sistemática y numéricamente predecible a las interacciones de la comunidad.

## Diagnóstico y Selección de Problema
A partir del análisis exploratorio, se confirma que este dataset corresponde a un conjunto de datos altamente fragmentado y con importantes desafíos de calidad, caracterizado por una alta concentración de contenido orientado al nicho de viajes y turismo (tags dominantes como thailand, travel, style) y una fuerte presencia del canal tecnológico Mrwhosetheboss. Estructuralmente, el dataset exhibe anomalías que requieren depuración, como una duplicidad de registros del 3.1% (581 IDs únicos frente a 600 filas totales) y un vacío de información crítico en sus metadatos, donde el 22.6% de los videos carece de contexto textual al registrar el string por defecto "No description available".  

Las variables numéricas presentan asimetrías positivas severas impulsadas por fenómenos estructurales de viralidad atípica (donde el 14% de la muestra califica como outlier en comentarios y un 13.33% supera las 14.8M de vistas). No obstante, la matriz de correlación de Spearman demostró interacciones orgánicas sumamente robustas y consistentes, destacando una sinergia lineal muy fuerte entre view_count y like_count (0.854), y un acoplamiento predecible aunque más exigente hacia la sección de comentarios (view_count vs comment_count de 0.743 y like_count vs comment_count de 0.731).  

Frente a estas características, se plantea abordar un problema de Clasificación Binaria para la Optimización de Inversión y Predicción de Alta Tracción de Contenido. En el ecosistema de distribución digital, las agencias de marketing y productoras de medios necesitan identificar tempranamente qué videos lograrán romper la barrera del rendimiento común para convertirse en activos de alta difusión, justificando así la asignación de presupuestos publicitarios o campañas de optimización SEO. Basarse únicamente en la intuición o en categorías rígidas genera ineficiencias operativas y financieras. Se diseñará un modelo predictivo capaz de clasificar si un contenido ingresado a la plataforma —considerando su categoría temática formal y su duración— cuenta con la estructura óptima para consolidarse como un Contenido de Alta Tracción. Este modelo actuará como un filtro inteligente de inventario para la toma de decisiones comerciales, permitiendo a los estrategas aislar el ruido de los metadatos imperfectos y maximizar el retorno de inversión (ROI) al concentrar los recursos de distribución exclusivamente en las publicaciones con mayor probabilidad de éxito masivo.  

La variable objetivo (y) será alta_traccion (Variable categórica binaria sintética: 1 para videos cuyo volumen de interacciones combinadas de vistas y likes se posiciona por encima de la mediana del dataset, y 0 para videos de rendimiento estándar o bajo). Las variables predictoras (X) incluirán la componente categórica formal category_id, que será procesada con One-Hot Encoding para transformarla en columnas binarias legibles por el modelo y las componentes numéricas continuas duration_seconds, view_count, like_count y comment_count. Se excluirán los identificadores únicos de alta cardinalidad video_id y thumbnail por su nulo valor predictivo, así como las columnas de texto libre title, description y tags para proteger al algoritmo de la dispersión extrema y del sesgo provocado por el 22.6% de nulos ocultos.

### Desafíos
El desafío técnico de este dataset está en la gestión de la asimetría extrema, la presencia de valores duplicados y los outliers masivos para evitar el sobreajuste (data leakage). La gran dispersión en las métricas de interacción y la diferencia de magnitudes físicas con la duración obligará a implementar transformaciones sobre las variables cuantitativas para estabilizar la varianza antes del modelado.

El desafío metodológico exigirá no confiarse de la métrica de Accuracy global debido al desbalance intrínseco que imponen los videos virales sobre el comportamiento típico de la muestra. Se priorizará la optimización del F1-Score y el análisis del área bajo la curva ROC-AUC, garantizando un equilibrio riguroso entre la Precisión y el Recall, para asegurar que el algoritmo no ignore aquellos contenidos con potencial real de despegue, consolidando así una herramienta analítica altamente confiable para la toma de decisiones comerciales.
