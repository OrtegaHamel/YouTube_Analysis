Adicionalmente, el valor mínimo de 0 en métricas de interacción (view_count, like_count) exigirá aplicar transformaciones logísticas o robustas antes del modelado.

```python
print("=== Coeficiente de Asimetría (Skewness) ===")
print(df[['view_count', 'like_count', 'comment_count', 'duration_seconds', 'engagement_rate']].skew())

columnas_num = ['view_count', 'like_count', 'comment_count', 'duration_seconds', 'engagement_rate']

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(14, 22))
fig.suptitle('Análisis de Distribución de Columnas Numéricas', fontsize=16, fontweight='bold', y=0.95)

sns.set_theme(style="whitegrid")

for i, col in enumerate(columnas_num):
    
    # Para evitar errores matemáticos con logaritmos si existen valores en 0, sumamos 1 de forma temporal al graficar
    if col in ['view_count', 'like_count', 'comment_count']:
        # --- Histograma con Escala Logarítmica para mitigar el sesgo de los outliers ---
        sns.histplot(data=df, x=df[col] + 1, kde=True, ax=axes[i, 0], color='crimson', bins=30, log_scale=True)
        axes[i, 0].set_title(f'Distribución de {col} (Escala Log)', fontsize=12, fontweight='semibold')
    else:
        # --- Histograma Lineal Estándar para Duración y Engagement ---
        sns.histplot(data=df, x=col, kde=True, ax=axes[i, 0], color='crimson', bins=30)
        axes[i, 0].set_title(f'Distribución de {col}', fontsize=12, fontweight='semibold')
        
    axes[i, 0].set_xlabel('')
    axes[i, 0].set_ylabel('Frecuencia')
    
    # --- Columna Derecha: Boxplot ---
    sns.boxplot(data=df, x=col, ax=axes[i, 1], color='salmon')
    axes[i, 1].set_title(f'Diagrama de Caja de {col}', fontsize=12, fontweight='semibold')
    axes[i, 1].set_xlabel('')

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()

```

Todas las variables numéricas, salvo exhiben coeficientes de asimetría (skewness) positivos, destacando de forma crítica la columna comment_count con un valor extremo de 20.78, seguida por like_count (5.83) y view_count (4.42). En los histogramas se ven distribuciones sesgadas a la derecha, donde la enorme masa de los datos se agrupa en valores bajos y una minoría de videos virales (outliers) muestra magnitudes millonarias. Asimismo, el factor de asimetría en duration_seconds (4.34) visualiza un ecosistema dominado por contenidos de corta duración (como Shorts), pero distorsionado en los diagramas de caja por la presencia de videos anómalos de extensa duración. Investigaremos más en profundidad esta columna.   
Este escenario de dispersión orgánica requerirá obligatoriamente el uso de escalas logarítmicas o transformaciones robustas para estabilizar la varianza antes de alimentar cualquier modelo predictivo.

```python
# Crear franjas de duración personalizadas para el análisis
def definir_franja(segundos):
    if segundos <= 5: return '1. Ultra-cortos (0-5s)'
    elif segundos <= 10: return '2. Rango Activo Inferior (6-10s)'
    elif segundos <= 30: return '3. Shorts Estándar (11-30s)'
    elif segundos <= 60: return '4. Límite de Shorts (31-60s)'
    else: return '5. Formato Largo Tradicional (>60s)'

df['Franja_Duracion'] = df['duration_seconds'].apply(definir_franja)

# Calcular el Máximo y la Mediana de vistas por cada franja
resumen_franjas = df.groupby('Franja_Duracion')['view_count'].agg(['count', 'median', 'max']).reset_index()
resumen_franjas['median'] = resumen_franjas['median'].apply(lambda x: f"{x:,.0f}")
resumen_franjas['max'] = resumen_franjas['max'].apply(lambda x: f"{x:,.0f}")

print("=== Análisis de Viralidad por Franjas de Duración ===")
print(resumen_franjas.to_string(index=False))
```

### Análisis de Correlación

```python

# Seleccionar solo las columnas numéricas con varianza (excluyendo favorite_count)
columnas_corr = [
    'view_count', 'like_count', 'comment_count', 
    'duration_seconds', 'engagement_rate', 
    'likes_to_views_ratio', 'comments_to_views_ratio', 'video_age_days'
]

matriz_corr = df[columnas_corr].corr(method='spearman')

print("=== Relaciones clave de correlación (Spearman) ===")
print(matriz_corr.unstack().sort_values(ascending=False).drop_duplicates().head(10))

plt.figure(figsize=(10, 8))

sns.heatmap(
    matriz_corr,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    vmin=-1, vmax=1,
    linewidths=0.5,
    cbar_kws={"label": "Coeficiente de Correlación (Spearman)"}
)

plt.title('Mapa de Calor de Correlación de Variables Numéricas', fontsize=14, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

El análisis de correlación revela que las métricas de volumen exhiben una fuerte sinergia positiva entre vistas y likes (0.83), aunque esta relación se debilita al interactuar con los comentarios (0.43). Asimismo, se evidencia que el aumento masivo de reproducciones reduce drásticamente las tasas de interacción, reflejado en correlaciones negativas entre view_count y los ratios de likes (-0.30) y comentarios (-0.54). Por otra parte, la correlación perfecta (1.00) entre engagement_rate y likes_to_views_ratio evidencia una redundancia algorítmica directa dentro del dataset.  
Los contenidos de mayor duración demuestran ser un formato de rendimiento de largo plazo, asociándose positivamente con una mayor antigüedad del video (0.36) y con la generación de un canal de debate más activo reflejado en la tasa de comentarios (0.51).

## Diagnóstico y Selección de Problema
A partir del análisis exploratorio, se confirma que corresponde a un conjunto de datos orgánicos y altamente especializados en el nicho de la tecnología. Estructuralmente presenta un estado limpio, pero sus distribuciones mostrando asimetrías positivas severas y la presencia de outliers masivos, conviviendo contenidos breves con megatutoriales de programación de alta duración. La matriz de correlación de Spearman demostró interacciones orgánicas verdaderas, exhibiendo una fuerte sinergia lineal entre vistas y likes (0.83), un efecto de penalización por viralidad en las tasas de enganche y una redundancia matemática perfecta (1.00) entre el ratio de likes y la tasa de interacción.

Frente a estas características, se plantea abordar un problema de Clasificación Binaria para la Optimización de Producción y Distribución de Contenido. No todo el contenido extenso es apto para ser picado en formatos cortos (Shorts), ya que algunos temas complejos requieren explicaciones pausadas y pierden el sentido si se fragmentan, resultando en un bajo engagement. Para una agencia de marketing digital o productora de medios, asignar a un editor a buscar fragmentos a ciegas en videos largos es operativamente costoso e ineficiente. Se diseñará un modelo predictivo capaz de clasificar si un contenido pilar de formato largo cuenta con la estructura e interacciones iniciales óptimas para ser fragmentado con éxito en piezas cortas de alto rendimiento. Este modelo actúa como un clasificador de inventario que indica al productor exactamente en qué contenidos enfocar los recursos de edición, maximizando el retorno de inversión de las horas-hombre y acelerando la estrategia de crecimiento multiplataforma.

La variable objetivo (y) será apto_para_shorts (Variable categórica binaria sintética con valores: 1 para alto potencial de fragmentación y 0 para formato exclusivamente largo). Las variables predictoras (X) incluirán la componente categórica category_id (excluyendo definition debido a su nula varianza estadística) y las componentes numéricas duration_seconds, view_count, like_count, comment_count, engagement_rate, likes_to_views_ratio, comments_to_views_ratio y video_age_days (excluyendo favorite_count por ser una constante).

### Desafíos
El desafío técnico de este dataset está en la severa dispersión e irregularidad diagnosticada en el EDA mediante la presencia extrema de outliers y desbalances críticos de clases. La diferencia de magnitudes físicas obligará a aplicar transformaciones robustas o logarítmicas y técnicas de escalamiento para la preparación de los datos antes del modelado. El desafío metodológico exigirá no confiarse de la métrica de Accuracy global. Se priorizará optimizar la precisión y la sensibilidad (Recall) mediante el análisis de curvas ROC-AUC, garantizando que el modelo minimice los falsos negativos (videos con alto potencial de fragmentación que el algoritmo no logra detectar) para asegurar una herramienta de auditoría de catálogo confiable en la toma de decisiones comerciales.
