# YouTube Video Statistics – Trending & Engagement
**Descripción**  
Métricas de rendimiento e interacción (engagement) para videos en tendencia de YouTube. Este conjunto de datos contiene 5.000 registros correspondientes a videos que alcanzaron a calificar como tendencias, distribuidos en diversas categorías como Música, Videojuegos (Gaming), Educación y Entretenimiento, entre otras. ### Análisis distribución de variables numéricas

```python
# Generamos estadísticas descriptivas del Data Frame
print("=" * 50)
print("Estadísticas descriptivas del Data Frame: ")
print("Columnas numéricas:")
display(df.describe())
```

Se observa que las métricas (views, likes, comments) tienen desvíos estándar controlados y una mediana muy cercana a la media, lo que sugiere distribuciones relativamente simétricas (poca presencia de outliers extremos).

```python
columnas_num = ['views', 'likes', 'comments', 'duration_seconds', 'engagement_rate']

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(14, 20))
fig.suptitle('Análisis de Distribución de Columnas Numéricas', fontsize=16, fontweight='bold', y=0.94)

sns.set_theme(style="whitegrid")

for i, col in enumerate(columnas_num):
    sns.histplot(data=df, x=col, kde=True, ax=axes[i, 0], color='royalblue', bins=30)
    axes[i, 0].set_title(f'Distribución de {col.capitalize()}', fontsize=12, fontweight='semibold')
    axes[i, 0].set_xlabel('')
    axes[i, 0].set_ylabel('Frecuencia')
    sns.boxplot(data=df, x=col, ax=axes[i, 1], color='lightskyblue')
    axes[i, 1].set_title(f'Boxplot de {col.capitalize()}', fontsize=12, fontweight='semibold')
    axes[i, 1].set_xlabel('')

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

Se puede ver una simetría en la mayoría de los histogramas, principalmente los de de views, likes y comments, donde las curvas son bastante acampanadas y uniformes. Esto confirma visualmente que la media y la mediana están muy cerca, lo que es un comportamiento típico de datos simulados o controlados.

En los boxplots de la derecha, no se distinguen puntos aislados o outliers fuera de los bigotes del gráfico. Esto reafirma que es un dataset sumamente limpio, ideal para entrenar un modelo lineal o un árbol sin necesidad de hacer una limpieza agresiva de datos atípicos.

### Análisis de Correlación

```python
columnas_num = ['views', 'likes', 'comments', 'duration_seconds', 'engagement_rate']
matriz_corr = df[columnas_num].corr()

plt.figure(figsize=(10, 8))
plt.title('Mapa de Calor de Correlación de Variables Numéricas', fontsize=14, fontweight='bold', pad=20)

sns.heatmap(matriz_corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```


Sorprendentemente, todos los coeficientes de correlación se sitúan en un rango cercano a cero (entre -0.034 y +0.023). Esto demuestra una ausencia total de relación lineal entre las métricas de rendimiento del video.
Esto podría levantar una alerta de un comportamiento anómalo de interacción, ya que en plataformas, las variables de volumen como views y likes presentan una co-linealidad positiva fuerte. En este dataset, la correlación es prácticamente nula (r = 0.0084), lo que rompe la lógica orgánica de embudos de conversión (vistas en interacciones).
Desde una perspectiva estrictamente de modelado, la ventaja de estos resultados es que los algoritmos no sufrirán distorsiones por redundancia de información entre predictores numéricos.



## Diagnóstico y Selección de Problema
A partir del Análisis Exploratorio de Datos (EDA) inicial, se sospecha que el Dataset corresponde a un conjunto de datos sintético o artificialmente controlado. Se presenta estructuralmente limpio (con 0 registros duplicados y 0 valores nulos), pero el análisis de la matriz de correlación reveló que todas las variables numéricas (views, likes, comments, duration_seconds y engagement_rate) presentan una independencia estadística absoluta entre sí, mostrando coeficientes de correlación prácticamente nulos. La ruptura de esta lógica en el dataset confirma su naturaleza artificial o preprocesada, aislando el comportamiento de las métricas numéricas.  

Frente a las características del conjunto de datos, se plantea abordar un problema de Clasificación Binaria.  
Se podría diseñar un modelo predictivo capaz de determinar si un video de la plataforma logrará posicionarse en la pestaña de tendencias globales. Para los creadores de contenido y las agencias de marketing digital, entender los factores asociados a la viralidad es un activo estratégico de alto valor. Un modelo que clasifique con precisión el potencial de un video permite optimizar los recursos de producción y pauta publicitaria antes o durante las primeras horas del lanzamiento, maximizando de forma científica el retorno de inversión (ROI) en plataformas digitales.  

**Variable Objetivo (y):** is_trending (Variable categórica binaria con valores: Yes / No).  
**Variables Predictoras o Características (X):**  
Componentes Categóricos: category y country.  
Componentes Numéricos: duration_seconds, views, likes, comments, engagement_rate.  

### Desafíos
El desafío técnico de este dataset radica en la estructura interna de los datos diagnosticada en el EDA: Ausencia de Señales Lineales y Dependencia de Variables Categóricas y Alta Cardinalidad.  
Debido a la naturaleza del problema (detección de videos virales), el foco no estará únicamente en la métrica de Accuracy global. El desafío metodológico exigirá optimizar la sensibilidad (Recall) y analizar exhaustivamente las curvas ROC-AUC, garantizando que el modelo minimice los falsos negativos (videos con potencial de tendencia que el algoritmo no logra detectar) y maneje correctamente cualquier desbalance previo en la variable is_trending.
